```python
from keras.models import load_model
import os
import cv2
import numpy as np
```

```python
model=load_model('emotion-059.keras')

face_clsf=cv2.CascadeClassifier(r'C:\Users\HP\Downloads\haarcascade_frontalface_default (1).xml')

source=cv2.VideoCapture(0)

labels_dict={0:'Angry',1:'Disgust',2:'Fear',3:'Happy',4:'Neutral',5:'Sad',6:'Surprise'}
color_dict={0:(0,0,255),1:(255,0,0),2:(128,128,128),3:(0,225,0),4:(0,255,255),5:(255,255,0),6:(0,215,255)}
```

```python
while(True):
    ret,img=source.read()
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_clsf.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)

    for (x,y,w,h) in faces:
        face_img=gray[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(100,100))
        normalized=resized/255.0
        reshaped=np.reshape(normalized,(1,100,100,1))
        result=model.predict(reshaped)

        label=np.argmax(result,axis=1)[0]

        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)
        cv2.rectangle(img,(x,y-50),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x+5, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
        
    cv2.imshow('Face Emotion Detection',img)
    key=cv2.waitKey(1)
    
    if(key==27):
        break
        
cv2.destroyAllWindows()
source.release()
```
